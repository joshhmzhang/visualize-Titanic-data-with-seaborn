```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic=pd.read_csv('train.csv',index_col=0)
titanic=titanic[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].dropna()
titanic.head()
```

```python
titanic['Age'].plot.hist(bins=20)
```

```python
sns.distplot(titanic['Age'])
```

```python
sns.kdeplot(titanic['Age'],shade=True,legend=False)
plt.xlabel('Age')
sns.despine()
sns.set_style('whitegrid')
```

Use a FacetGrid instance to generate three plots on the same row

```python
g=sns.FacetGrid(titanic,col='Pclass',height=4)
g.map(sns.kdeplot,'Age',shade=True)
```

```python
g=sns.FacetGrid(titanic,col='Pclass',row='Survived',height=4)
g.map(sns.kdeplot,'Age',shade=True)
```

```python
g=sns.FacetGrid(titanic,col='Pclass',row='Survived',hue='Sex',height=4)
g.map(sns.kdeplot,'Age',shade=True)
g.add_legend(fontsize=18)
plt.show()
```
